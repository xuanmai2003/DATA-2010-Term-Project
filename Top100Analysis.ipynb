```python
import pandas as pd

df = pd.read_csv("songs_after_2010_2.csv")

df.head()
```

```python
is_number1 = df["peak_position"] == 1
number1_no_duplicates = df[is_number1].groupby("song").first().sort_values("week_id", ascending = True)
number1_no_duplicates.sample(5)
```

```python
top_num = 15

graph_data = number1_no_duplicates["performer"].value_counts().head(top_num)
graph_data.sample(10)
```

```python
df2 = pd.read_csv('song_after_2010.csv')

df2 = df2.drop("url",1)
df2 = df2.drop("instance",1)
df2 = df2.drop("previous_week_position",1)
df2 = df2.drop("weeks_on_chart",1)
df2 = df2.drop("spotify_genre",1)
df2 = df2.drop("spotify_track_id",1)
df2 = df2.drop("spotify_track_preview_url",1)
df2 = df2.drop("spotify_track_duration_ms",1)
df2 = df2.drop("spotify_track_explicit",1)
df2 = df2.drop("spotify_track_album",1)
df2 = df2.drop("week_position",1)
#df2 = df2.drop("performer",1)
df2 = df2.drop("song_id",1)
df2 = df2.iloc[: , 1:]

df2 = df2.fillna(0)

is_number1_2 = df2["peak_position"] == 1

number1_no_duplicates2 = df2[is_number1_2].groupby("song").first().sort_values("week_id", ascending = True)
number1_no_duplicates2 = number1_no_duplicates2.drop("week_id",1)
number1_no_duplicates2 = number1_no_duplicates2.drop("peak_position",1)

number1_no_duplicates2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X = number1_no_duplicates2.drop('performer', axis = 1)
y = number1_no_duplicates2.performer

# We specify the random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
(X_train.shape, X_test.shape)

model = KNeighborsClassifier(n_neighbors = 5)
model.fit(X_train, y_train)

# Get predicted values
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier(criterion = 'entropy')
model2.fit(X_train, y_train)

# Get predicted values
y_pred2 = model.predict(X_test)
y_pred2
```

```python
print((X_train.shape, X_test.shape))

print(model2.get_depth())

print(model2.get_n_leaves())
```

```python
from matplotlib import pyplot as plt
from sklearn import tree

# Make the figure larger
plt.rcParams['figure.figsize'] = [15, 5]

# Specify the feature names using the column names of X_train
tree.plot_tree(model2, max_depth = 3,
               feature_names = list(X_train.columns))
plt.show()
```
